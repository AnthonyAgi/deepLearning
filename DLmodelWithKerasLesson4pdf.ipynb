```python
import pandas as pd

df1 = pd.read_csv('diabetes.csv')
df1.head()
```

```python
x = df1.iloc[:, 0:8]
y = df1.iloc[:, 8]

x.head()
```

```python
y.head()
```

```python
df1.describe()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
m1 = Sequential()
m1.add(Dense(12, input_dim=8, activation='relu'))
m1.add(Dense(8, activation='relu'))
m1.add(Dense(1, activation='sigmoid'))
```

```python
m1.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
m1.fit(x, y, epochs=60, batch_size=10)
```

```python
_, accuracy = m1.evaluate(x, y)
print ('Accuracy: %.2f' % (accuracy*100))
```
